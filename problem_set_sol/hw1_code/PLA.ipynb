```python
iter_num = []
for k in range(1000):
    data_10 = np.random.rand(100,2)*2-1

    data_10.reshape(100,2)

    label_10 = []
    for i in data_10:
        if i[0]>=i[1]:
            label_10.append(1)
        else:
            label_10.append(-1)

    label_10 = np.array(label_10)
    
    clf = Perceptron(tol=1e-3)
    print(label_10)
    clf.fit(data_10,label_10)

    iter_num.append(clf.n_iter_)
print(iter_num)
print(np.mean(np.array(iter_num)))
```
